```python
%run utility.ipynb
import pandas as pd
import altair as alt


def df():
    df = loadAndCleanData("1201Data-Anonymized.csv")
    #df = df.replace("--",0.0)
    pd.set_option('display.max_columns', None)
    return(df)
```

```python
z = df()
```

```python
#DataFrame 1
    #This is a dataframe consisting of the students who had completed late work
def df1():
    hw1 = z.loc[z["Quiz: Homework 1 LATE WORK (Real)"] != 0]
    hw1 = hw1.copy()
    hw2 = z.loc[z["Quiz: Homework 2 LATE WORK (Real)"] != 0]
    hw2 = hw2.copy()
    hw3 = z.loc[z["Quiz: Homework 3 LATE WORK (Real)"] != 0]
    hw3 = hw3.copy()
    hw4 = z.loc[z["Quiz: Homework 4 LATE WORK (Real)"] != 0]
    hw4 = hw4.copy()
    hw5 = z.loc[z["Quiz: Homework 5 LATE WORK (Real)"] != 0]
    hw5 = hw5.copy()
    hw6 = z.loc[z["Quiz: Homework 6 LATE WORK (Real)"] != 0]
    hw6 = hw6.copy()
    hw7 = z.loc[z["Quiz: Homework 7 LATE WORK (Real)"] != 0]
    hw7 = hw7.copy()
    hw8 = z.loc[z["Homework 8 Late Actual"] != 0]
    hw8 = hw8.copy()
    
    lateDF = hw1
    lateDF = lateDF.append([hw2,hw3,hw4,hw5,hw6,hw7,hw8])
    
    lateDF = lateDF.drop_duplicates(subset="UID")
    
    
    
    return(lateDF)
```

```python
df1()
```

```python
def avg1():
    lateDF = df1()
    lateDF["Final Grade"] = lateDF["Final Grade"].astype(int)
    sum1 = lateDF["Final Grade"].sum()
    len1 = len(lateDF["Final Grade"])
    #print(sum1)
    #print(len1)


    #This is the average grade of the students who completed late work
    avg = sum1 / len1
    #return(lateDF)
    return(avg)
```

```python
avg1()
```

```python
#DataFrame 2
    #This is a dataframe consisting of the students who had completed no late work
def df2():
    z = df()
    
    z["Quiz: Homework 1 LATE WORK (Real)"] = z["Quiz: Homework 1 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 2 LATE WORK (Real)"] = z["Quiz: Homework 2 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 3 LATE WORK (Real)"] = z["Quiz: Homework 3 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 4 LATE WORK (Real)"] = z["Quiz: Homework 4 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 5 LATE WORK (Real)"] = z["Quiz: Homework 5 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 6 LATE WORK (Real)"] = z["Quiz: Homework 6 LATE WORK (Real)"].astype(float)
    z["Quiz: Homework 7 LATE WORK (Real)"] = z["Quiz: Homework 7 LATE WORK (Real)"].astype(float)
    z["Homework 8 Late Actual"] = z["Homework 8 Late Actual"].astype(float)
    
    
    newDF = z.loc[z["Quiz: Homework 3 LATE WORK (Real)"] == 0.000]
    newDF1 = newDF[(z["Quiz: Homework 1 LATE WORK (Real)"] == 0.0) & (z["Quiz: Homework 2 LATE WORK (Real)"] == 0.0)
                  & (z["Quiz: Homework 3 LATE WORK (Real)"] == 0.000) & (z["Quiz: Homework 4 LATE WORK (Real)"] == 0.0)
                   & (z["Quiz: Homework 5 LATE WORK (Real)"] == 0.0) & (z["Quiz: Homework 6 LATE WORK (Real)"] == 0.0)
                   & (z["Quiz: Homework 7 LATE WORK (Real)"] == 0.0) & (z["Homework 8 Late Actual"] == 0.0)
                  ]

    
    
    pd.set_option('display.max_rows', None)
    return(newDF1)
```

```python
df2()
```

```python
def avg2():
    lateDF2 = df2()
    lateDF2["Final Grade"] = lateDF2["Final Grade"].astype(int)
    sum1 = lateDF2["Final Grade"].sum()
    len1 = len(lateDF2["Final Grade"])
    #print(sum1)
    #print(len1)


    #This is the average grade of the students who completed late work
    avg = sum1 / len1
    return(avg)
```
