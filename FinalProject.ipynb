### Creating Data Subsets

DATAFRAME 1
<br>*This is a dataframe consisting of the students who had completed late work.*

```python
# create empty dataframe
dfLate = pd.DataFrame(columns=order)
```

```python
# populate empty dataframe
for feature in df.columns:
    if '_L' in feature and 'Lab' not in feature:
        if dfLate.size == 0:
            dfLate = df[df[feature] != 0]
        else:
            dfLate = dfLate.append(df[df[feature] != 0])
```

```python
# clean it up a bit more
dfLate.sort_index(inplace=True)
dfLate.drop_duplicates(inplace=True)
```

```python
dfLate
```

DATAFRAME 2
<br>*This is a dataframe consisting of the students who had completed no late work.*

```python
# create empty dataframe
dfNoLate = pd.DataFrame(columns=order)
```

```python
# populate empty dataframe
for feature in df.columns:
    if '_L' in feature and 'Lab' not in feature:
        if dfNoLate.size == 0:
            dfNoLate = df[df[feature] == 0]
        else:
            dfNoLate = dfNoLate.append(df[df[feature] == 0])
```

```python
dfNoLate = dfNoLate[(dfNoLate["Q_Hwk1_L"] == 0_0) & (dfNoLate["Q_Hwk2_L"] == 0_0) & (dfNoLate["Q_Hwk3_L"] == 0_000) &
                    (dfNoLate["Q_Hwk4_L"] == 0_0) & (dfNoLate["Q_Hwk5_L"] == 0_0) & (dfNoLate["Q_Hwk6_L"] == 0_0) &
                    (dfNoLate["Q_Hwk7_L"] == 0_0) & (dfNoLate["Q_Hwk8_L"] == 0_0)]
```

```python
# clean it up a bit more
dfNoLate.sort_index(inplace=True)
dfNoLate.drop_duplicates(inplace=True)
```

### Visualizations

```python
x = dfLate.reset_index()

xrange = (round(dfLate["Final_Grade"].min())-10, round(dfLate["Final_Grade"].max())+10)

scatter2 = alt.Chart(x).mark_point().encode(
    x = alt.X("UID", title=""),
    y = alt.Y("Final_Grade", title="Final Grade", scale=alt.Scale(domain=xrange)),
    tooltip = ["Final_Grade"]
)

rule2 = alt.Chart(dfLate).mark_rule(color="red").encode(
    y = alt.Y("mean(Final_Grade)", title="Final Grade"),
    tooltip = ["mean(Final_Grade)"]
)

(scatter2 + rule2).properties(width=650)
```

```python
x = dfNoLate.reset_index()

xrange = (round(dfNoLate["Final_Grade"].min())-10, round(dfNoLate["Final_Grade"].max())+10)

scatter2 = alt.Chart(x).mark_point().encode(
    x = alt.X("UID", title=""),
    y = alt.Y("Final_Grade", title="Final Grade", scale=alt.Scale(domain=xrange)),
    color = alt.Color("Final_Grade", legend=alt.Legend(title="Final Grade")),
    tooltip = ["Final_Grade"]
)

rule2 = alt.Chart(dfNoLate).mark_rule(color="red").encode(
    y = alt.Y("mean(Final_Grade)", title="Final Grade"),
    tooltip = ["mean(Final_Grade)"]
)

(scatter2 + rule2)
```
